```python
## Missing Value Assessment

### Observation

The dataset contains a total of **11 missing values**, all of which are present in the **Postal Code** column.

### Analysis

Postal Code is a geographical identifier and is not a critical field for the business analysis performed in this project. The missing values do not affect the calculation of key performance indicators such as Sales, Profit, Quantity, Customer Segmentation, or Regional Analysis.

### Decision

The missing values will be retained because removing the corresponding rows would result in unnecessary loss of valid business transactions. The Postal Code column will not be used for KPI calculations or dashboard development.
```

```python
## Data Type Validation

### Observation

The dataset was inspected to verify that each column had an appropriate data type.

### Findings

- Numerical columns such as Sales, Profit, Discount, and Quantity were already stored using appropriate numeric data types.
- Order Date and Ship Date were initially stored as datetime for time analysis
```

```python
# Create a new column for the order year

df["Year"] = df["Order Date"].dt.year
df[["Order Date", "Year"]].head()
```

```python
# Create a new column for the month number

df["Month"] = df["Order Date"].dt.month
df[["Order Date", "Month"]].head()
```

```python
# Create a new column for the month name

df["Month Name"] = df["Order Date"].dt.month_name()
df[["Order Date", "Month Name"]].head()
```

```python
# Create a new column for the quarter

df["Quarter"] = df["Order Date"].dt.quarter
df[["Order Date", "Quarter"]].head()
```

```python
# Create a new column for the day of the week

df["Day of Week"] = df["Order Date"].dt.day_name()
df[["Order Date", "Day of Week"]].head()
```

```python
## Feature Engineering

### Objective

Feature engineering was performed to derive additional business-friendly attributes from the Order Date column.

### New Features Created

- Year
- Month
- Month Name
- Quarter
- Day of Week

### Business Value

These derived features simplify trend analysis and will be used throughout the project for Python analysis, SQL KPI calculations, Power BI dashboard development, and AI-generated business insights.
```

```python
# Calculate total sales by category

sales_by_category = (
    df.groupby("Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

sales_by_category
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot Sales by Category

plt.figure(figsize=(8, 5))

sales_by_category.plot(kind="bar")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.xticks(rotation=0)

plt.show()
```

```python
## Business Insight – Sales by Category

### Observation

Technology generated the highest total sales among all product categories, followed by Furniture and Office Supplies.

### Business Impact

Technology products are the primary revenue driver for the business. The company should ensure adequate inventory, promotional campaigns, and product availability for this category while exploring opportunities to improve sales in the remaining categories.
```

```python
# Calculate total sales by region

sales_by_region = (
    df.groupby("Region")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

sales_by_region
```

```python
# Plot Sales by Region

plt.figure(figsize=(8,5))

sales_by_region.plot(kind="bar")

plt.title("Total Sales by Region")

plt.xlabel("Region")

plt.ylabel("Total Sales")

plt.xticks(rotation=0)

plt.show()
```

```python
## Business Insight – Sales by Region

### Observation

The West region generated the highest total sales, while the South region recorded the lowest sales.

### Business Impact

The company should continue strengthening its presence in the West region while investigating the factors contributing to lower sales in the South region. Potential strategies include targeted marketing campaigns, inventory optimization, and customer acquisition initiatives.
```

```python
# Calculate monthly sales trend

monthly_sales = (
    df.groupby(["Year", "Month"])["Sales"]
      .sum()
      .reset_index()
)

monthly_sales.head()
```

```python
# Create a Year-Month date column

monthly_sales["Year_Month"] = pd.to_datetime(
    monthly_sales["Year"].astype(str) + "-" +
    monthly_sales["Month"].astype(str)
)
```

```python
# Plot Monthly Sales Trend

plt.figure(figsize=(12,6))

plt.plot(
    monthly_sales["Year_Month"],
    monthly_sales["Sales"],
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")

plt.grid(True)

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
## Business Insight – Monthly Sales Trend

### Observation

Monthly sales fluctuate throughout the four-year period, indicating seasonal variations and changing customer demand. Certain months show noticeable sales peaks, while others experience relatively lower performance.

### Business Impact

The company can use these trends for demand forecasting, inventory planning, and scheduling promotional campaigns during slower sales periods.
```
